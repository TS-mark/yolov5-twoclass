```python
"""
@description  :图片转存
---------
@param  :
    path:待转存目录
    new_path:转存目录
    file_type:转存的文件类型
-------
@Returns  :
    info：转存成功返回"done"
-------
"""
def CopyFile(path, new_path, file_type = ""):
    if not os.path.exists(new_path):
        os.makedirs(new_path)

    for ipath in os.listdir(path):
        if file_type and ipath.endswith("."+file_type):
            fulldir = os.path.join(path, ipath)  # 拼接成绝对路径
            print(fulldir)   # 打印相关后缀的文件路径及名称
            if os.path.isfile(fulldir):  # 文件，匹配->打印
                shutil.copy(fulldir,new_path)

            if os.path.isdir(fulldir):  # 目录，递归   
                FindFile(fulldir)

                
    return "done"
```

```python
"""
@description  :删除路径下的文件
---------
@param  :
    path:待删除路径
    file_type:待删除的文件类型
-------
@Returns  :
    info：成功返回"done"，否则返回对应问题信息
-------
"""
def RemoveFile(path, file_type = ""):
    if not os.path.exists(path):
        return "File path not exists!"
    
    if not os.path.isdir(path):
        return r"it`s not the file path！"
    
    for f in os.listdir(path):
        if file_type and f.endswith(file_type):
            print("remove file" + os.path.join(path,f))
            os.remove(os.path.join(path,f))
    
    return "done"


"""
@description  :保留路径下的某种类型文件，其余全部删除
---------
@param  :
    path:待删除路径
    file_type:保存的文件类型，其余全部删除
-------
@Returns  :
    info：成功返回"done"，否则返回对应问题信息
-------
"""
def onlyRemainFile(path, file_type = ""):
    if not os.path.exists(path):
        return "File path not exists!"
    
    if not os.path.isdir(path):
        return r"it`s not the file path！"
    
    for f in os.listdir(path):
        if file_type and not f.endswith(file_type):
            print("remove file" + os.path.join(path,f))
            os.remove(os.path.join(path,f))
            break
        
    
    return "done"
```

```python
label_path1 = "./data/annotation_txt/train"
label_path2 = "./data/annotation_txt/test"
tar_dir = r"C:\python project\datasets\tt100K\labels\train"
```

```python
def make_YOLOv5_label(labelpath, tarpath):
    if not os.path.exists(tarpath):
        os.makedirs(tarpath)
    flist = os.listdir(labelpath)
    for tx in flist:
        wr_res = []
        with open(os.path.join(labelpath, tx), "r") as f1:
            for i in f1.readlines():
                li = i.split("\t")
                if li[0] == "lz1" or li[0] == "lz2":
                    # li[0] = "lb2"
                    li[0] = "2"
                else:
                    # li[0] = "lb1"
                    li[0] = "1"
                wr_res.append("\t".join(li))

        with open(os.path.join(tarpath, tx), "w") as f2:
            f2.write("".join(wr_res))
        # break
```

```python
make_YOLOv5_label(label_path2, tar_dir)
make_YOLOv5_label(label_path1, tar_dir)
```
