```python
import os
import shutil
```

```python
path = r".\enhanced"
dir_path = r".\classifier_tt100k_2021"
file_list = os.listdir(path)
```

```python
file_list
```

```python
# 移动一个文件夹的文件
for file in file_list:
    temp_f = os.listdir(path + r"/" + file)
    for f in temp_f:
        # if f.endswith(".txt"):
        # print("ori",path + r"/" + file + r"/" + f)
        # print("dir",dir_path + r"/" + file + r"/" + f)
        shutil.move(path + r"/" + file + r"/" + f,
                dir_path + r"/" + file + r"/" + f)
            
```
